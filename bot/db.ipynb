```python
import psycopg2 as pg
import os
from dotenv import load_dotenv

load_dotenv()

con = pg.connect(dbname="QuantumBot",user=os.getenv("PG_USERNAME"),password=os.getenv("PG_PASSWORD"),host="localhost")
if not con: raise Exception("Connection cannot create")
```

```python
def clean(values): return clean(values[0]) if len(values) < 2 else values
```

```python
cursor = con.cursor()
cursor.execute("select weight from ticker where ticker = 'AMD'")
row = cursor.fetchall()
print(len(row))
row = clean(row)
row[0] # just weights
```
